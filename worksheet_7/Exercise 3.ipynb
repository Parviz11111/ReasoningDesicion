```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load Examples dataset
examples1 = pd.read_csv('Example1.csv')
examples2 = pd.read_csv('Example2.csv')
examples3 = pd.read_csv('Example3.csv')

# Split Examples dataset into training and test sets
examples1_train, examples1_test = train_test_split(examples1, test_size=0.2, random_state=42)
examples2_train, examples2_test = train_test_split(examples2, test_size=0.2, random_state=42)
examples3_train, examples3_test = train_test_split(examples3, test_size=0.2, random_state=42)

# Load Iris dataset
iris = pd.read_csv('iris-numclass.csv')

# Split Iris dataset into training and test sets
iris_train, iris_test = train_test_split(iris, test_size=0.2, random_state=42)
```

```python
from scipy.stats import multivariate_normal

def calculate_posterior_probability(feature, mean, cov):
    return multivariate_normal.pdf(feature, mean=mean, cov=cov)

def classify_examples(features, class_params):
    predictions = []
    for feature in features:
        max_posterior = 0
        predicted_class = None
        for class_label, (mean, cov) in class_params.items():
            posterior = calculate_posterior_probability(feature, mean, cov)
            if posterior > max_posterior:
                max_posterior = posterior
                predicted_class = class_label
        predictions.append(predicted_class)
    return predictions

# Prepare training data for Examples dataset
examples_train = {
    'Example1': examples1_train.iloc[:, 1:],
    'Example2': examples2_train.iloc[:, 1:],
    'Example3': examples3_train.iloc[:, 1:]
}

# Prepare test data for Examples dataset
examples_test = {
    'Example1': examples1_test.iloc[:, 1:],
    'Example2': examples2_test.iloc[:, 1:],
    'Example3': examples3_test.iloc[:, 1:]
}

# Estimate parameters for each class in Examples dataset
class_params = {}
for class_label, class_data in examples_train.items():
    mean = class_data.mean().values
    cov = class_data.cov().values
    class_params[class_label] = (mean, cov)

# Classify test data for Examples dataset
examples_predictions = {}
for class_label, test_data in examples_test.items():
    predictions = classify_examples(test_data.values, class_params)
    examples_predictions[class_label] = predictions

# Print classification results for Examples dataset
for class_label, predictions in examples_predictions.items():
    print(f"Classification results for {class_label}:")
    print(predictions)
    print()
```
