```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Example1.csv
df1 = pd.read_csv('Example1.csv', header=None)
class_labels1 = df1.iloc[0]
features1 = df1.iloc[1:]

# Load Example2.csv
df2 = pd.read_csv('Example2.csv', header=None)
class_labels2 = df2.iloc[0]
features2 = df2.iloc[1:]

# Load Example3.csv
df3 = pd.read_csv('Example3.csv', header=None)
class_labels3 = df3.iloc[0]
features3 = df3.iloc[1:]

# Plot Example1.csv
plt.scatter(features1.iloc[0], features1.iloc[1], c=class_labels1)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Example1.csv')
plt.show()

# Plot Example2.csv
plt.scatter(features2.iloc[0], features2.iloc[1], c=class_labels2)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Example2.csv')
plt.show()

# Plot Example3.csv
plt.scatter(features3.iloc[0], features3.iloc[1], c=class_labels3)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Example3.csv')
plt.show()
```

```python
# Load iris-numclass.csv
iris_df = pd.read_csv('iris-numclass.csv', header=None)
class_labels_iris = iris_df.iloc[:, 0]
iris_features = iris_df.iloc[:, 1:]

# Plotting features in different combinations
combinations = [(0, 1), (0, 2), (0, 3), (1, 2), (1, 3), (2, 3)]

for feature_indices in combinations:
    plt.scatter(iris_features.iloc[:, feature_indices[0]], iris_features.iloc[:, feature_indices[1]], c=class_labels_iris)
    plt.xlabel(f'Feature {feature_indices[0]+1}')
    plt.ylabel(f'Feature {feature_indices[1]+1}')
    plt.title(f'Iris Dataset (Features {feature_indices[0]+1} vs {feature_indices[1]+1})')
    plt.show()
```
