### Normally distributed multivariate random numbers ###
#### a) Implement a way to generate normally distributed random vectors of dimensions 2, 5 and 10 with mean vector µ = 0 and covariance matrix Σ = I. ####

```python
import numpy as np


# Generate random vectors for each dimension
def normallyDistributedVector(dim, sampleSize):

    mean = np.zeros(dim)    # Create a mean vector filled with zeros with dimension dim

    cov = np.eye(dim)       # Create a covariance matrix as identity matrix with dimension dim

    samples = np.random.multivariate_normal(mean, cov, sampleSize)

    
    return(samples)

Set1 = normallyDistributedVector(2, 10)
Set2 = normallyDistributedVector(5, 10) 
Set3 = normallyDistributedVector(10, 10) 

print(f"Random vectors of dimension 2:\n{Set1}\n")
print(f"Random vectors of dimension 5:\n{Set2}\n")
print(f"Random vectors of dimension 10:\n{Set3}\n")
```

#### b) Use the random vector generator from 1 a) to create a sample set of random vectors. Estimate the parameters of the underlying normal distribution from this sample set. ####
#### What do you observe? How accurate is your parameter estimate? What happens if the size of the sample set changes? ####



```python
Set1 = normallyDistributedVector(5, 10) 
Set2 = normallyDistributedVector(5, 1000)
Set3 = normallyDistributedVector(5, 100000)


# Estimate the mean and covariance matrix of the underlying distribution
sample_mean = np.mean(Set1, axis=0)
sample_cov = np.cov(Set1.T)

print(f"SampleSize: 10\n")
print(f"Sample mean:\n{sample_mean}\n")
print(f"Sample covariance matrix:\n{sample_cov}\n")

sample_mean = np.mean(Set2, axis=0)
sample_cov = np.cov(Set2.T)

print(f"SampleSize: 1000\n")
print(f"Sample mean:\n{sample_mean}\n")
print(f"Sample covariance matrix:\n{sample_cov}\n")

sample_mean = np.mean(Set3, axis=0)
sample_cov = np.cov(Set3.T)

print(f"SampleSize: 100000\n")
print(f"Sample mean:\n{sample_mean}\n")
print(f"Sample covariance matrix:\n{sample_cov}\n")
```
